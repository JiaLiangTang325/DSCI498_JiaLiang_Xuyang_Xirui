```python
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import os
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
pipe.enable_attention_slicing()
```

```python
content_folder = "C:/Users/26911/Desktop/diffusion_style_project/data/content"
output_folder = "C:/Users/26911/Desktop/diffusion_style_project/results"

os.makedirs(output_folder, exist_ok=True)
```

```python
styles = {
    "vangogh": "a painting in the style of Vincent van Gogh",
    "monet": "a painting in the style of Claude Monet",
    "cubism": "cubism painting style, by Pablo Picasso"
}
```

```python
image_files = os.listdir(content_folder)

print("Total images:", len(image_files))
```

```python
for img_name in image_files:

    img_path = os.path.join(content_folder, img_name)

    init_image = Image.open(img_path).convert("RGB")
    init_image = init_image.resize((512, 512))

    for style_name, prompt in styles.items():

        result = pipe(
            prompt=prompt,
            image=init_image,
            strength=0.75,
            guidance_scale=7.5
        )

        output_image = result.images[0]

        save_name = f"{img_name.split('.')[0]}_{style_name}.png"
        save_path = os.path.join(output_folder, save_name)

        output_image.save(save_path)

        print("Saved:", save_name)
```
